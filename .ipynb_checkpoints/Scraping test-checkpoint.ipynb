```python
# Test scraping of dn.no

def retrive_names_and_links():
    driver = webdriver.Chrome(ChromeDriverManager().install())
    links_names = []
    links = []
    names = []

    for i in tqdm(range(1, 100)): # Change the limit of the range
        url = f'https://www.dn.no/smak/vinsok?available_cust=&p={i}'
        driver.get(url)
        time.sleep(2)
        htmlSource = driver.page_source
        link_locations = htmlSource.split('href="')[1:]
        for link in link_locations:
            link = link.split('</a>')[0]
            links_names.append(link)
        links_names = [link for link in links_names if '/vin/' in link] 
        links_names = list(dict.fromkeys(links_names))
        for link in links_names:
            split = link.split('>')[0]
            name = link.split('>')[1]
            links.append(split)
            names.append(name)
    return links, names
```

```python
links_names = retrive_names_and_links()
```

```python
links = links_names[0]
names = links_names[1]

print(len(links))
print(len(names))

# Remove duplicates
#links = list(dict.fromkeys(links))
#names = list(dict.fromkeys(names))

print(len(links))
print(len(names))
```

```python
df_dn = pd.DataFrame([])
df_dn['name'] = names
df_dn['link'] = links
df_dn.head()
```

```python
# Test scrape by searching names on Vivino
base_url = 'vivino.com'

search_name = []
for name in names: 
    search_name.append(str(name).replace(' ', '+'))
    
driver = webdriver.Chrome(ChromeDriverManager().install())

vivino_link_locations = []
vivino_links_names = []
vivino_links = []
vivino_names = []

for name in tqdm(search_name[0:250]): 
    search_url = f'https://www.vivino.com/search/wines?q={name}'
    driver.get(search_url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source,'html')
    vivino_links_names = soup.find_all('span', {'class':'header-smaller text-block wine-card__name'})

    for x in vivino_links_names: 
        vivino_names.append(x.text.replace('\n', ''))

    for x in vivino_links_names: 
        temp = str(str(x).split('href=')[1]).split('><')[0].replace('"', '')
        vivino_links.append(base_url + temp)
```

```python
print('Names: \n', vivino_names)
print('Links: \n', vivino_links)
```

```python
df_vivino = pd.DataFrame([])
df_vivino['vivino_name'] = vivino_names
df_vivino['vivino_link'] = vivino_links
df_vivino
```

```python
df_vivino_copy = df_vivino.copy()
df_dn_copy = df_dn.copy()
df_vivino_copy.columns = ['name', 'link']
```

```python
df_merged = pd.merge(df_dn_copy, df_vivino_copy, how='left', on='name')
df_merged
```
